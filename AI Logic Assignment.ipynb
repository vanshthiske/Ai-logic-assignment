 ## Problem statement 1 

```python
def mearge_interval(intervals):
    intervals.sort(key=lambda x:x[0])

    merged =[]

    for interval in intervals:
        if not merged or merged [-1][1]<interval[0]:
            merged.append(interval)
        else:
            merged[-1][1]=max(merged[-1][1],interval[1])
    return merged
        
```

```python
def test_case():
    n = int(input().strip())
    
    intervals =[]
    
    for _ in range(n):
        start, end =map(int, input().split())
        intervals.append([start, end])
    result = mearge_interval(intervals)
    return result
```

## Problem Statement 2 

```python
def addition():
    n = int(input().strip())
    arr = list(map(int, input().split()))
    k = int(input().strip())

    w = sum(arr[:k])
    max_sum=w

    for i in range(k,n):
        w+=arr[i]
        w-=arr[i-k]
        max_sum=max(max_sum, w)

    print(max_sum)
```

## Problem statement 3 

```python
def Channel_Analyzer():
    s = input().strip()
    char_set= set()
    left = 0
    max_len= 0 

    for i in range(len(s)):
        while s[i] in char_set:
            char_set.remove(s[left])
            left+=1

        char_set.add(s[i])
        max_len=max(max_len,i- left+1)
    return max_len
        
```
